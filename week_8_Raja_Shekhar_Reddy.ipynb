# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
import logging

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s | %(levelname)s | %(message)s"
)
````

````python
# 🤖 AGENT FUNCTION (TO IMPLEMENT)

def agent(query: str):
    query_lower = query.lower()

    if "calculate" in query_lower:
            expression = query_lower.replace("calculate", "").strip()

            return {
                "type": "calculation",
                "result": calculator(expression)
            }

        # Route to Keyword Extractor
    elif "keywords" in query_lower:
            text = query.replace("Extract keywords from", "")
            text = text.replace("extract keywords from", "").strip()

            return {
                "type": "keywords",
                "result": extract_keywords(text)
            }
    else:
      return {
                "type": "general",
                "result": "I'm a smart assistant. Please ask me to calculate something or extract keywords."
            }
````

# Bonus tasks

````python
import logging
from datetime import datetime

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s | %(levelname)s | %(message)s"
)

def word_count(text):
    return len(text.split())

def reverse_text(text):
    return text[::-1]

def current_datetime():
    return datetime.now().strftime("%Y-%m-%d %H:%M:%S")

def agent(query: str):

    try:

        logging.info(f"Received Query: {query}")

        query_lower = query.lower().strip()

        if any(word in query_lower for word in
               ["calculate", "add", "sum", "subtract",
                "multiply", "divide"]):

            expression = query_lower

            for word in ["calculate", "add", "sum",
                         "subtract", "multiply", "divide"]:
                expression = expression.replace(word, "")

            expression = expression.strip()

            logging.info("Routing to Calculator Tool")

            return {
                "type": "calculation",
                "result": calculator(expression)
            }

        elif any(word in query_lower for word in
                 ["keyword", "keywords", "extract"]):

            text = query
            text = text.replace("Extract keywords from", "")
            text = text.replace("extract keywords from", "")
            text = text.replace("extract", "")
            text = text.replace("keywords", "")

            logging.info("Routing to Keyword Tool")

            return {
                "type": "keywords",
                "result": extract_keywords(text.strip())
            }

        elif "word count" in query_lower:

            text = query.replace("word count", "").strip()

            logging.info("Routing to Word Count Tool")

            return {
                "type": "word_count",
                "result": word_count(text)
            }

        elif "reverse" in query_lower:

            text = query.replace("reverse", "").strip()

            logging.info("Routing to Reverse Tool")

            return {
                "type": "reverse",
                "result": reverse_text(text)
            }

        elif "time" in query_lower or "date" in query_lower:

            logging.info("Routing to Date-Time Tool")

            return {
                "type": "datetime",
                "result": current_datetime()
            }

        else:

            logging.info("General Query")

            return {
                "type": "general",
                "result": "I'm a smart assistant. I can calculate expressions, extract keywords, count words, reverse text, and tell the current date & time."
            }

    except Exception as e:

        logging.error(f"Error: {str(e)}")

        return {
            "type": "error",
            "result": str(e)
        }
````

````python

while True:
    query = input("Enter query (type 'exit' to stop): ")
    if query.lower() == "exit":
        break
    print(agent(query))
````
